```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')

plt.rcParams['figure.figsize'] = (10, 5)
```

```python
bikes = pd.read_csv(r'C:\Users\Lenovo\Downloads\bikes - Sheet1.csv', encoding='utf-8', parse_dates=['Date'], dayfirst=True, index_col='Date')
bikes['Berri 1'].plot()
```

```python
berri_bikes = bikes[['Berri 1']].copy()
```

```python
berri_bikes[:5]
```

```python
berri_bikes.index.weekday
```

```python
berri_bikes.loc[:,'Weekday'] = berri_bikes.index.weekday
berri_bikes[:5]
```

```python
weekday_counts = berri_bikes.groupby('Weekday').aggregate(sum)
weekday_counts
```

```python
weekday_counts.index = ['Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Sunday']
weekday_counts
```

```python
weekday_counts.plot(kind='bar')
```
